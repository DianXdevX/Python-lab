## Introdução às bibliotecas de dados

Python possui MUITAS bibliotecas para trabalhar com dados.

As três bibliotecas mais utilizadas são:

- **NumPy** → computação numérica e arrays
- **Pandas** → manipulação de dados em tabelas
- **Matplotlib** → criação de gráficos

```python
# importando bibliotecas

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## O que é NumPy?

NumPy é uma biblioteca usada para trabalhar com **arrays e operações matemáticas eficientes**.

Ela é muito mais rápida que listas comuns do Python para operações numéricas.

```python
import numpy as np

array = np.array([1, 2, 3, 4, 5])

print(array)
```

## O que é Pandas?

Pandas é usado para manipular **dados estruturados**, como tabelas.

A principal estrutura de dados é o **DataFrame**, que funciona como uma planilha.

```python
import pandas as pd

dados = {
    "nome": ["Ana", "Carlos", "Maria"],
    "idade": [25, 30, 28]
}

df = pd.DataFrame(dados)

print(df)
```

## O que é Matplotlib?

Matplotlib é uma biblioteca usada para **criar gráficos e visualizações de dados**.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y)
plt.show()
```

# Manipulação de dados

## Arrays com NumPy

Arrays são estruturas usadas para armazenar **vários valores numéricos** de forma eficiente.


```python
import numpy as np

numeros = np.array([10, 20, 30, 40])

print(numeros)
```

```python
# operações com arrays

numeros = np.array([1,2,3,4])

resultado = numeros * 2

print(resultado)
```

## DataFrames com Pandas

DataFrames representam **tabelas de dados**.

```python
import pandas as pd

dados = {
    "produto": ["Notebook", "Mouse", "Teclado"],
    "preco": [3500, 120, 200]
}

df = pd.DataFrame(dados)

print(df)
```

## Leitura de dados

O Pandas permite ler arquivos como:

- CSV
- Excel
- JSON

```python
# Criar o arquivo CSV

import pandas as pd

dados = {
    "produto": ["Notebook", "Mouse", "Teclado"],
    "preco": [3500, 120, 200]
}

df = pd.DataFrame(dados)

df.to_csv("dados.csv", index=False)
```

```python
import pandas as pd

df = pd.read_csv("dados.csv")

print(df)
```

## Operações básicas

Podemos acessar colunas e realizar cálculos.

```python
print(df["preco"])
```

## Operações estatísticas

```python
df["preco"].mean()
```

```python
df["produto"].count()
```

```python
df.describe()
```

## Agrupamentos

Podemos agrupar dados por categorias.

```python
dados = {
    "produto": ["Notebook","Mouse","Mouse","Teclado"],
    "preco": [3500,120,150,200]
}

df = pd.DataFrame(dados)

df.groupby("produto").mean()
```

# Visualização com matplotlib

```python
# Gráfico de linhas

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,15,20,30]

plt.plot(x,y)
plt.show()
```

```python
# Gráfico de barras

produtos = ["Notebook","Mouse","Teclado"]
vendas = [15,50,30]

plt.bar(produtos,vendas)
plt.show()
```

```python
# Gráfico de dispersão

x = [1,2,3,4,5]
y = [2,4,5,4,5]

plt.scatter(x,y)
plt.show()
```

# Integração das bibliotecas

Na prática, usamos **NumPy, Pandas e Matplotlib juntos**.

Fluxo típico de análise de dados:

1️⃣ ler dados  
2️⃣ processar dados  
3️⃣ analisar dados  
4️⃣ visualizar dados

```python
import pandas as pd
import matplotlib.pyplot as plt

dados = {
    "produto": ["Notebook","Mouse","Teclado"],
    "vendas": [15,50,30]
}

df = pd.DataFrame(dados)

plt.bar(df["produto"], df["vendas"])

plt.show()
```
